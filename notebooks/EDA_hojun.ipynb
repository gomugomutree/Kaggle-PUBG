```python
def drow_boxplot(feature):
    plt.figure(figsize=(20, 12))
    plt.subplot(211)
    sns.boxplot(data=df, x=feature, y='winPlacePerc')
    plt.subplot(212)
    sns.countplot(data=df, x=feature)
    plt.show()
```

```python
def merge_count_rate(feature):
    return pd.merge(df[feature].value_counts().round(3), df[feature].value_counts(normalize = True).round(3), right_index = True,
               left_index = True)
```

```python
import math

def make_subplot_layout(df):
    col_num = len(df)
    k = len(df.columns)
    row_num = math.ceil(k/col_num)
    plt.figure(figsize=(col_num*4,row_num*3))
    for i in range(k) : 
        plt.subplot(row_num,col_num,i+1) #  +1  .
        sns.boxenplot(x=df.iloc[:,i], y='winPlacePerc', data=df)
        plt.title(df.columns[i])
    plt.tight_layout()
    plt.show()
```
